```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("data.csv",index_col=0)
```

```python
data
```

```python
emotion=data['情感倾向'].sort_values(ascending=False)
good=data['好评'].sort_values(ascending=False)
neutual=data['中评'].sort_values(ascending=False)
bad=data['差评'].sort_values(ascending=False)
total=data['总评论数'].sort_values(ascending=False)
pic=data['有图'].sort_values(ascending=False)
```

```python
plt.rcParams['font.sans-serif']=['SimHei']#显示中文标签 
plt.rcParams['axes.unicode_minus']=False
fig = plt.figure(figsize=(20, 5))
plt.xticks(rotation=30)
plt.tick_params(labelsize=16)
plt.title('情感倾向',size=20)
plt.bar(emotion.index.values,emotion.values,color='r',alpha=0.5,label="情感")
plt.legend()
```

```python
plt.rcParams['font.sans-serif']=['SimHei']#显示中文标签 
plt.rcParams['axes.unicode_minus']=False
fig = plt.figure(figsize=(20, 5))
plt.xticks(rotation=30)
plt.tick_params(labelsize=16)
plt.title('总评论数',size=20)
plt.bar(total.index.values,total.values,color='r',alpha=0.5,label="总评")
plt.legend()
```

```python
plt.rcParams['font.sans-serif']=['SimHei']#显示中文标签 
plt.rcParams['axes.unicode_minus']=False
fig = plt.figure(figsize=(20, 5))
plt.xticks(rotation=30)
plt.tick_params(labelsize=16)
plt.title('差评',size=20)
plt.bar(bad.index.values,bad.values,color='g',alpha=0.5,label="差评")
plt.legend()
```

```python
plt.rcParams['font.sans-serif']=['SimHei']#显示中文标签 
plt.rcParams['axes.unicode_minus']=False
fig = plt.figure(figsize=(20, 5))
plt.xticks(rotation=30)
plt.tick_params(labelsize=16)
plt.title('好评',size=20)
plt.bar(good.index.values,good.values,color='b',alpha=0.5,label="好评")
plt.legend()
```

```python
data['好评率']=data['好评']/data['总评论数']
data['差评率']=data['差评']/data['总评论数']
```

```python
good_rate=data['好评率'].sort_values(ascending=False)
bad_rate=data['差评率'].sort_values(ascending=False)
```

```python
plt.rcParams['font.sans-serif']=['SimHei']#显示中文标签 
plt.rcParams['axes.unicode_minus']=False
fig = plt.figure(figsize=(20, 5))
plt.xticks(rotation=30)
plt.tick_params(labelsize=16)
plt.title('好评率',size=20)
plt.bar(good_rate.index.values,good_rate.values,color='b',alpha=0.5,label="好评率")
plt.legend()
```

```python
plt.rcParams['font.sans-serif']=['SimHei']#显示中文标签 
plt.rcParams['axes.unicode_minus']=False
fig = plt.figure(figsize=(20, 5))
plt.xticks(rotation=30)
plt.tick_params(labelsize=16)
plt.title('差评率',size=20)
plt.bar(bad_rate.index.values,bad_rate.values,color='b',alpha=0.5,label="好评率")
plt.legend()
```

```python
cmap = sns.diverging_palette(200,20,sep=20,as_cmap=True)
sns.heatmap(data.corr(),cmap=cmap)
plt.title('特征相关性热力图')
plt.show()
```
